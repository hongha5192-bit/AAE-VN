Notebook này chạy pipeline setup + train từ dataset `giaphlm/aae-new`.

```python
import subprocess
from pathlib import Path

candidates = [
    Path('/kaggle/input/datasets/giaphlm/aae-new/AAE-new/deploy/v2/kaggle_kernel/run_kernel.py'),
    Path('/kaggle/input/aae-new/AAE-new/deploy/v2/kaggle_kernel/run_kernel.py'),
]
for p in Path('/kaggle/input').rglob('run_kernel.py'):
    if 'deploy/v2/kaggle_kernel' in str(p):
        candidates.append(p)

script = next((p for p in candidates if p.exists()), None)
if script is None:
    raise FileNotFoundError('Cannot find deploy/v2/kaggle_kernel/run_kernel.py under /kaggle/input')

print('Using', script)
subprocess.run(['python3', '-u', str(script)], check=True)
```
